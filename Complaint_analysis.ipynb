```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("complaints.csv", usecols=["Issue"], nrows=50000)
```

```python
top_issues = df["Issue"].value_counts().head(10)
```

```python
print(top_issues)
```

```python
top_issues.plot(kind="bar")
plt.title("Top 10 Complaint Issues")
plt.xlabel("Issue")
plt.ylabel("Count")
plt.savefig("cpmplaint_issue_chart.png")
plt.show()
```

```python
df = pd.read_csv("complaints.csv", usecols=["Submitted via"], nrows=50000)
channel_counts = df["Submitted via"].value_counts().head(10)
print(channel_counts)
```

```python
channel_counts.plot(kind="bar")
plt.title("Complaints by submission channel")
plt.xlabel("Channel")
plt.ylabel("Count")
plt.savefig("Submission_channel_chart.png")
plt.show()
```

```python
df = pd.read_csv("complaints.csv", usecols=["Company"], nrows=50000)
top_companies = df["Company"].value_counts().head(10)
print(top_companies)
```
